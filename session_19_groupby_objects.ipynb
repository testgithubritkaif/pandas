```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/content/imdb-top-1000.csv')
```

```python
movies.head()
```

```python
genres=movies.groupby('Genre')
```

```python
# Applying builtin aggregation fuctions on groupby objects
genres.min()
genres.max()
genres.mean(numeric_only=True)
genres.median(numeric_only=True)
genres.std(numeric_only=True)
genres.count()
genres.sum()
```

```python
# find the top 3 genres by total earning
movies.groupby('Genre')["Gross"].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')["IMDB_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
# agg method
# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list
numeric_cols = genres.obj.select_dtypes(include='number').columns
genres[numeric_cols].agg(['min','max','mean','sum'])
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
max_imdb_movies = []
for group,data in genres:
  max_imdb_movies.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

df = pd.concat(max_imdb_movies)
df
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
# find number of movies starting with A for each group
def str_A(A):
  return A["Series_Title"].str.startswith('A').sum()
genres.apply(str_A)
```

```python
# find ranking of each movie in the group according to IMDB scor

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group["normalized_rating"] = (group["IMDB_Rating"] - group["IMDB_Rating"].min()) / (group["IMDB_Rating"].max() - group["IMDB_Rating"].min())
  return group

genres.apply(normal)
```

```python
   # groupby on multiple cols
duo = movies.groupby(['Director','Star1'])
duo
# # size
duo.size()
# # get_group
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find the most earning actor->director combo
movies.groupby(['Director','Star1'])["Gross"].sum().sort_values(ascending=False).head(1)
# duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
movies
movies.groupby(["Star1","Genre"])["Metascore"].mean().reset_index().sort_values("Metascore",ascending=False)
```

```python
# agg on multiple groupby
aggrega = duo.obj.select_dtypes(include='number').columns
duo[aggrega].agg(['min','max','mean'])
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head()
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10)
ipl.groupby("batsman").apply(lambda x:x["batsman_runs"].sum()).sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
ipl.groupby("batsman")["total_runs"].apply(lambda x: (x==6).sum()).sort_values(ascending=False).head(1)
# six = ipl[ipl['batsman_runs'] == 6]
# six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
df = ipl[ipl["over"]>15]
# df.groupby("batsman").apply(lambda x: (x["batsman_runs"]==4).sum() + (x["batsman_runs"]==6).sum()).sort_values(ascending=False)head(1)
df = df[(df["batsman_runs"]==4) | (df["batsman_runs"]==6)]
df.groupby("batsman")["batsman"].count().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
df = ipl[ipl["batsman"]=="V Kohli"]
df.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False).reset_index()
```

```python
# Create a function that can return the highest score of any batsman

def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]

highest('DA Warner')
```
