###`Q-5:` In IPL matches dataset some teams name has changed.
You will have to consider them as same.
```
'Delhi Capitals' formerly as 'Delhi Daredevils'
'Punjab Kings' formerly as 'Kings XI Punjab'
'Rising Pune Supergiant' formerly as 'Rising Pune Supergiants'
```
You need to make changes accordingly. Consider current name for each teams.

Be careful Gujrat Titans and Gujrat Lions are different teams.

````python
# code here
ipl_match = pd.read_csv("/content/IPL_Matches_2008_2022.csv")
ipl_match.head()
````

````python
#  'Delhi Capitals' formerly as 'Delhi Daredevils'
# 'Punjab Kings' formerly as 'Kings XI Punjab'
# 'Rising Pune Supergiant' formerly as 'Rising Pune Supergiants'

ipl_match.replace({'Delhi Daredevils':'Delhi Capitals' ,'Kings XI Punjab':'Punjab Kings','Rising Pune Supergiants':'Rising Pune Supergiant'})

ipl_match.rename(columns={'Delhi Daredevils':'Delhi Capitals' ,'Kings XI Punjab':'Punjab Kings','Rising Pune Supergiants':'Rising Pune Supergiant'}, inplace=True)
ipl_match.sample(30)
````

###`Q-6` Write a code which can display the bar chart of top 5 teams who have played maximum number of matches in the IPL.
>  Hint: Be careful the data is divided in 2 different cols(Team 1 and Team 2)


````python
# code here
# Write a code which can display the bar chart of top 5 teams who have played maximum number of matches in the IPL.
import matplotlib.pyplot as plt
maixmum_team_play = pd.concat([ipl_match["Team1"], ipl_match["Team2"]]).value_counts()
print(maixmum_team_play)
maixmum_team_play.head(5).plot(kind="bar")
plt.title('Top 5 Teams by Number of Matches Played')
plt.xlabel('Team')
plt.ylabel('Number of Matches')
````

###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```


````python
ipl_match.head()
````

````python
ipl_match.head()
def team1_vs_team2(team1, team2):
  team_win = ipl_match[(ipl_match["Team1"]== team1) &  (ipl_match["Team2"]==team2)]
  team_win = team_win["WinningTeam"].value_counts().sort_values(ascending=False).head(1)
  pl_of_match = ipl_match[(ipl_match["Team1"]== team1) &  (ipl_match["Team2"]==team2)]
  pl_of_match = pl_of_match["Player_of_Match"].value_counts().sort_values(ascending=False).head(1)
  print(pl_of_match)
  print(team_win)

team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
````

### `Q-9:` Find out the top 7 cities where the matches of Kolkata Knight Riders are played frequently and plot the result as bar chart.

*`.plot(kind = "bar")` can help you to plot the bar chart. Also you can learn more about this method from [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)*



````python
# code here
# Find out the top 7 cities where the matches of Kolkata Knight Riders are played frequently and plot the result as bar chart
import matplotlib.pyplot as plt
df = ipl_match[(ipl_match["Team1"] == "Kolkata Knight Riders") | (ipl_match["Team2"] == "Kolkata Knight Riders")]
df = df["City"].value_counts().sort_values(ascending=False).head(7)
df.plot(kind="bar")
plt.title('Top 7 Cities by Number of Matches Played')
plt.xlabel('City')
plt.ylabel('Number of Matches')


````

### `Q-10:` Find out the average margin for the team Mumbai Indians for only the session 2011.

````python
# code here
avg_margin = ipl_match[(ipl_match["Season"] == "2011") & ((ipl_match["Team1"] == "Mumbai Indians") | (ipl_match["Team2"] == "Mumbai Indians"))]
avg_margin = avg_margin["Margin"].mean()
print(f" the average margin for the team Mumbai Indians for only the session 2011 = {avg_margin}")
````
